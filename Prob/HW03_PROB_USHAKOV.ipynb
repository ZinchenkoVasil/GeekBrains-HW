1. Даны значения зарплат из выборки выпускников: 100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150. Посчитать (желательно без использования статистических методов наподобие std, var, mean) среднее арифметическое, среднее квадратичное отклонение, смещенную и несмещенную оценки дисперсий для данной выборки.

```python
data = np.array([100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150])
data_mean = data.sum() / data.shape[0]
print("Cреднее арифметическое:", data_mean)
data_sq_diff = sum((data - data_mean) ** 2)
print("Cреднеквадратичное отклонение:", (data_sq_diff/data.shape[0]) ** 0.5)
data_disp = data_sq_diff / (data.shape[0]-1)
print("Оценка дисперсии для выборки (смещенная):", data_disp)
print("Оценка дисперсии как для генеральной совокупности (несмещенная):", data_sq_diff/data.shape[0])
```

4. В университет на факультеты A и B поступило равное количество студентов, а на факультет C студентов поступило столько же, сколько на A и B вместе. Вероятность того, что студент факультета A сдаст первую сессию, равна 0.8. Для студента факультета B эта вероятность равна 0.7, а для студента факультета C - 0.9. Студент сдал первую сессию. Какова вероятность, что он учится: a). на факультете A б). на факультете B в). на факультете C?

```python
P = (1/4 * 0.8 + 1/4 * 0.7 + 1/2 *0.9)
print("Априорная вероятность того что выбранный случайным образом студент здаст первую сессию:",P)
print("Вероятность того что студент здавший сессию учится на факультете А:", 1/4 * 0.8 / P)
print("Вероятность того что студент здавший сессию учится на факультете B:", 1/4 * 0.7 / P)
print("Вероятность того что студент здавший сессию учится на факультете C:", 1/2 * 0.9 / P)
```
